```python
def price_per_day(days, curr):
    days_info = cryptocompare.get_historical_price_day(curr, currency='USD', limit=days)
    return [(day_info['high']+day_info['low'])/2 for day_info in days_info]
```

```python
DAYS = 2000

btc2usd = price_per_day(DAYS, 'BTC')
eth2usd = price_per_day(DAYS, 'ETH')

f'Last {round(DAYS/365, 2)} years'
```

```python
eth_btc_ratio = [(eth/btc)*100 for eth, btc in zip(eth2usd, btc2usd)]
plt.plot([i for i in range(DAYS+1)], eth_btc_ratio)
```
