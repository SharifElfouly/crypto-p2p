```python
from web3 import Web3
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
plt = plt.plot
```

```python
w3 = Web3(Web3.HTTPProvider('https://mainnet.infura.io/v3/786a7764b8234b06b4cd6764a1646a17'))
w3.isConnected()
```

```python
def get_block(n):     return w3.eth.get_block(n)
def get_block_txs(n): return len(get_block(n)['transactions'])
```

```python
n_blocks = get_block('latest')['number']
'{:_}'.format(n_blocks) + ' blocks'
```

```python
EVERY_N_BLOCKS = 10_000

n_txs = [get_block_txs(i) for i in range(n_blocks, 0, -EVERY_N_BLOCKS)]
```

```python
x = [i for i in range(len(n_txs))]
plt(x, n_txs)
```
